```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import shutil
```

```python
maps_data = pd.DataFrame(columns = ['C1', 'C2','C3','C4','C5','audio_path', 'midi_path', 'csv_path', 'pred_midi_path'])

smd_data = pd.DataFrame(columns = ['C1', 'C2','C3','C4','C5','audio_path', 'midi_path', 'csv_path', 'pred_midi_path'])
```

```python
# For MAPS dataset
# Works with SMD too
flac_files = []
mid_files = []
txt_files = []
for root, dirs, files in os.walk("./SMD", topdown=False):
    for name in files:
        #print(name, root, dirs)
        # extract the extension of the file
        ext = name.split('.')[-1]
        # perform processing on string names so they are all similarly decorated
        # all files have '.' in them
        name_list =  name.split('.')[:-1]
        if len(name_list)>1:
            fname2 = '_'.join(name_list)
        else:
            fname2 = name_list
        if type(fname2) == list:
            fname2 = fname2[0]
        # some even has '-'
        name_list2 =  fname2.split('-')
        if len(name_list2)>1:
            fname = '_'.join(name_list2)
        else:
            fname = name_list2
        if type(fname) == list:
            fname = fname[0]
        #
        
        # save files in different lists based on file type        
        if ext == 'flac' or ext == 'wav' :
            # flac_files.append((root, fname,os.path.join(root, fname+ '.' +ext)))
            flac_files.append((root, fname, os.path.join(root, name)))
        elif ext == 'mid':
            mid_files.append((root, fname, os.path.join(root, name)))
        elif ext == 'txt':
            txt_files.append((root, fname, os.path.join(root, name)))
        else:
            # raise error
            print(name, ext)
            raise ValueError("Unsupported File")
```

```python
flac_files
```

```python
maps_path = './MAPS/preds/'
smd_path = './SMDpred/'
for idx, tupl in enumerate(flac_files):
    root,name,path = tupl
    #print(root, name, path)
    vals = root.split("/")[3:]
    pred_midi_path = os.path.join(smd_path + '/'.join(vals) +'/' + name + '.mid')
    # Maximum folder depth is 5
    while len(vals)!=5:
        vals.append('')
    # add audio
    vals.append(path)
    # find midi files
    for root2,name2,path2 in mid_files:
        if name == name2:
            vals.append(path2)
    # find csv files
    for root3,name3,path3 in txt_files:
        if name == name3:
            vals.append(path3)
    # for dataset without csv
    while len(vals)!=8:
        vals.append('')
    # add Prediction midi path
    vals.append(pred_midi_path)
    # pth = os.path.join()
    # print(vals)
    maps_data.loc[len(maps_data)] = vals
    if idx%6000 == 0:
        print(idx, vals)
        print(pred_midi_path, name, root)
        # print(root, name)
    # clears list for next iteration
    vals.clear()
```
