```python
def time_smoothing(arr, filter_ = np.array([1,1,1])):
    pad_len = (len(filter_)-1)//2
    pad_arr = np.pad(arr, pad_width=pad_len)
    new_arr = np.zeros_like(arr)
    deno_ = np.sum(filter_)
    for i in range(len(filter_)):
        new_arr[i] = np.sum(pad_arr[i:i+len(filter_)]*filter_) / deno_
    return new_arr  
    
```
